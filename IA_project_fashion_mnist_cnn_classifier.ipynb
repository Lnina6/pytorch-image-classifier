```python
import torchvision
import torchvision.transforms as transforms
import torch

train_dataset = torchvision.datasets.FashionMNIST( "data", train=True, download=True,
                                                  transform=transforms.ToTensor(),)

test_dataset = torchvision.datasets.FashionMNIST("data", train=False, download=True,
                                                  transform=transforms.ToTensor(),)
# Split train_dataset into a train and validation sets:
train_size = int(0.75*len(train_dataset))
valid_size = len(train_dataset) - train_size# Habituellement, les ensembles d'entraînement sont plus grands que les ensembles de validation.,
# Split train_dataset into train and validation sets
train_dataset, valid_dataset = torch.utils.data.random_split(train_dataset, [train_size, valid_size])
# Ground-Truth classes in the FashionMNIST:
classes = ('T-shirt/top', 'Trouser', 'Pullover', 'Dress',
           'Coat', '	Sandal', '	Shirt', 'Sneaker', 'Bag', 'Ankle boot')
```

```python
#DATA loading and visualisation
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

#DataLoader for the train_set (renvoie des batch de tensor )
train_loader=DataLoader(train_dataset, batch_size=36, shuffle=True)#les donnets, taille de chaque mini-lot d'images=batch=64 images,
#DataLoader for the test_set
test_loader=DataLoader(test_dataset, batch_size=36, shuffle=False)
#DataLoader for the valid_set
valid_loader=DataLoader(valid_dataset, batch_size=36, shuffle=False)

#function that takes the batches, as iput .outut void.show images and labels
def show_image_label(batch):
  # Display image and label.
  train_features, train_labels= next(iter(batch))
  print(f"Feature batch shape: {train_features.size()}")
  print(f"Labels batch shape: {train_labels.size()}")
  for i in range(len(train_features)):  # Loop through each image in the batch
        img = train_features[i]  # Extract the image
        label = train_labels[i]  # Extract the label

        # Permute dimensions of the image tensor for proper display
        img = img.permute((1, 2, 0))

        plt.figure(figsize=(6, 6))  # Adjust figure size
        plt.imshow(img.squeeze(), cmap='gray', interpolation='nearest')  # Display the image as grayscale
        plt.show()  # Show the plot
        print(f"Label: {label}")  # Print the label of the image
        print("Image shape:", img.shape)  # Print the shape of the image

print("The train set contains {} images, in {} batches".format(len(train_loader.dataset), len(train_loader)))
print("The validation set contains {} images, in {} batches".format(len(valid_loader.dataset), len(valid_loader)))
print("The test set contains {} images, in {} batches".format(len(test_loader.dataset), len(test_loader)))
show_image_label(train_loader)
```

```python
##Designing a Convolutional Neural Network
import torch
import torch.nn as nn
import torch.nn.functional as F

#defining une class CNN avec 2 couche de convolution qui renvoie la sortie final du reseau
#image 28*28*1, 3 couche de convolution?
#nn.Module: Base class for all neural network modules.

class SimpleCNN(nn.Module):

    def __init__(self):
        super(SimpleCNN, self).__init__()
        # Define convolutional layers
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=20, kernel_size=5, stride=1, padding=1)
        self.conv2 = nn.Conv2d(in_channels=20, out_channels=46, kernel_size=4, stride=1, padding=1)
        self.conv3 = nn.Conv2d(in_channels=46, out_channels=70, kernel_size=3, stride=1, padding=1)
        # Define max pooling layer
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        # Define fully connected layers
        self.fc1 = nn.Linear(70 * 3 * 3, 512)  # Adjust the size based on the input image size
        self.fc2 = nn.Linear(512, 10)  #

    def forward(self, x):
        # Apply convolutional and pooling layers
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        # Flatten the tensor for fully connected layers
        x = x.view(-1, 70 * 3 * 3)  # Adjust the size based on the input image size
        # Apply fully connected layers
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Instantiate the model
cnn_model = SimpleCNN()

# Print the model architecture
print(cnn_model)
```

```python
#Dimension of feature maps in CNN
#kernel=noyau(masque); stride=le pas; padding=process of addign layers(to the input)
def OutputTensor_size(input_size, kernel_size, padding_amount, stride):
    output_size=((input_size-kernel_size-(2*padding_amount))//stride)+1 #le 1 est la premiere valeur qu'on a
    return output_size
```

```python
#Loss function and optimizer
import torch.optim as optim
#The learning rate determines the size of the steps taken during optimization.
#Momentum is a parameter that accelerates optimization in the relevant direction and dampens oscillations.

# Using the  neural network architecture defined precedently
cnn_model = SimpleCNN()
# Loss function, critéere d'optimisation pour calculer la perte.utiliser pendant la phase test
loss = nn.CrossEntropyLoss()
#the optimizer
optimizer= optim.SGD(cnn_model.parameters(), lr=0.01, momentum=0.9)
```

```python
##Training Loop
from torch.utils.tensorboard import SummaryWriter
from sklearn.metrics import precision_score, recall_score, f1_score
import torch.nn.functional as F
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np

num_iterations=10
l2_lambda = 0.0005  # Example value, adjust as needed
writer = SummaryWriter()
#boocle d'entrainement:
#rajouts des tableau pour stocket les differentes métrique en focntion des epoches,modif 26-03
training_losses = []
validation_losses = []
accuracies = []
precisions = []
recalls = []
f1_scores = []
epochs = []
for epoch in range(num_iterations):
  #mettre le model en mode entrainment
  cnn_model.train()
  running_loss=0.0

  #ietration sur le datset d'entrainement
  for i, data in enumerate(train_loader):
    #data is a list
    inputs, labels=data
    #efacer les gradients precedament calculer
    optimizer.zero_grad()
    #passer les entréees par les difrentes couche de notre model, et obtenir notre prediction
    outputs=cnn_model(inputs)
    #calcule des erreurs entre le model et les donnets reel
    loss_value=loss(outputs, labels)

    # Apply L2 regularization
    l2_reg = sum(torch.sum(param ** 2) for param in cnn_model.parameters())
    loss_value += l2_lambda * l2_reg

    #calcule gradient
    loss_value.backward()
    #optimisation des poids, pour avancer dans la bonne direction
    optimizer.step()

    # Print statistics
    running_loss = running_loss+loss_value.item()
    # Write the scalar value of the loss to Tensorboard
    writer.add_scalar('training_loss', loss_value.item(), epoch * len(train_loader) + i)
    if i % 2000 == 1999:  # Print every 2000 mini-batches
      print('[%d, %5d] loss: %.3f' %
        (epoch + 1, i + 1, running_loss / 2000))
      running_loss = 0.0

# Validation
  cnn_model.eval()  # Set the model to evaluation mode
  val_loss = 0.0
  correct = 0
  total = 0
  predictions = []
  true_labels = []

  # Disable gradient computation during validation
  with torch.no_grad():
      for data in valid_loader:
          inputs, labels = data
          outputs = cnn_model(inputs)
          loss_value = loss(outputs, labels)
          val_loss += loss_value.item()

          # Get the predicted labels
          _, predicted = torch.max(outputs.data, 1)
          total += labels.size(0)
          correct += (predicted == labels).sum().item()

          predictions.extend(predicted.cpu().numpy())
          true_labels.extend(labels.cpu().numpy())

  # Calculate validation loss and accuracy and the other metrics
  val_loss /= len(valid_loader)
  accuracy = 100 * correct / total
  precision = precision_score(true_labels, predictions, average='macro')
  recall = recall_score(true_labels, predictions, average='macro')
  f1 = f1_score(true_labels, predictions, average='macro')

  # Ajoutez ces métriques aux listes respectives,modif 26-03
  training_losses.append(running_loss / len(train_loader))
  validation_losses.append(val_loss)
  accuracies.append(accuracy)
  precisions.append(precision)
  recalls.append(recall)
  f1_scores.append(f1)
  epochs.append(epoch + 1)

  print(f'Epoch [{epoch + 1}/{num_iterations}], Validation Loss: {val_loss:.3f}, Accuracy: {accuracy:.2f}%')
  print(f'Precision: {precision:.3f}, Recall: {recall:.3f}, F1 Score: {f1:.3f}')

  #,modif 26-03
  # À la fin de l'ensemble de validation
  if epoch == num_iterations - 1:
      # Tracez les graphiques jusqu'à cette époque
      plt.figure(figsize=(10, 5))

        # Loss curve
      plt.subplot(2, 2, 1)
      plt.plot(epochs, training_losses, label='Training Loss')
      plt.plot(epochs, validation_losses, label='Validation Loss')
      plt.xlabel('Epoch')
      plt.ylabel('Loss')
      plt.title('Training and Validation Loss')
      plt.legend()

        # Accuracy curve
      plt.subplot(2, 2, 2)
      plt.plot(epochs, accuracies, label='Accuracy', color='green')
      plt.xlabel('Epoch')
      plt.ylabel('Accuracy (%)')
      plt.title('Accuracy')
      plt.legend()

        # Precision, Recall, and F1 Score curves
      plt.subplot(2, 2, 3)
      plt.plot(epochs, precisions, label='Precision', color='orange')
      plt.plot(epochs, recalls, label='Recall', color='blue')
      plt.plot(epochs, f1_scores, label='F1 Score', color='red')
      plt.xlabel('Epoch')
      plt.ylabel('Score')
      plt.title('Precision, Recall, and F1 Score')
      plt.legend()

      plt.tight_layout()
      plt.show()


# Visualize confusion matrix
  conf_matrix = confusion_matrix(true_labels, predictions)
  fig, ax = plt.subplots(figsize=(8, 8))
  ax.matshow(conf_matrix, cmap='Blues')
  num_classes = len(classes)
  ax.set_xticks(np.arange(num_classes))
  ax.set_yticks(np.arange(num_classes))
  ax.set_xticklabels(classes)
  ax.set_yticklabels(classes)
  for i in range(num_classes):
      for j in range(num_classes):
          ax.text(j, i, str(conf_matrix[i, j]), ha='center', va='center')
  plt.xlabel('Predicted Label')
  plt.ylabel('True Label')
  plt.title('Confusion Matrix')
  plt.show()

  # Visualize misclassified images
  #misclassified_images = []
  #for i in range(len(true_labels)):
   #   if true_labels[i] != predictions[i]:
  #        misclassified_images.append((inputs[i], classes[true_labels[i]], classes[predictions[i]]))

  #fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))
  #for i, (image, true_label, predicted_label) in enumerate(misclassified_images[:9]):
   #   ax = axes[i // 3, i % 3]
    #  ax.imshow(image.squeeze().cpu().numpy(), cmap='gray')
     # ax.set_title(f'True: {true_label}, Predicted: {predicted_label}')
      #ax.axis('off')
  #plt.tight_layout()
  #plt.show()

print('Finished Training')
```

```python
#"showing the misqualified images with
##Training Loop
from torch.utils.tensorboard import SummaryWriter
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
import torch

num_iterations = 10
l2_lambda = 0.0005  # Example value, adjust as needed
writer = SummaryWriter()

# Placeholder arrays for tracking metrics
training_losses = []
validation_losses = []
accuracies = []
precisions = []
recalls = []
f1_scores = []
epochs = []
misclassified_images = []  # Store misclassified images

# Training loop
for epoch in range(num_iterations):
    cnn_model.train()
    running_loss = 0.0

    # Training phase
    for i, data in enumerate(train_loader):
        inputs, labels = data
        optimizer.zero_grad()
        outputs = cnn_model(inputs)
        loss_value = loss(outputs, labels)

        # Apply L2 regularization
        l2_reg = sum(torch.sum(param ** 2) for param in cnn_model.parameters())
        loss_value += l2_lambda * l2_reg

        loss_value.backward()
        optimizer.step()
        running_loss += loss_value.item()
        writer.add_scalar('training_loss', loss_value.item(), epoch * len(train_loader) + i)

    # Validation phase
    cnn_model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    predictions = []
    true_labels = []

    with torch.no_grad():
        for data in valid_loader:
            inputs, labels = data
            outputs = cnn_model(inputs)
            loss_value = loss(outputs, labels)
            val_loss += loss_value.item()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            predictions.extend(predicted.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

            # Collect misclassified images
            for i in range(inputs.size(0)):
                if predicted[i] != labels[i]:
                    misclassified_images.append((inputs[i], labels[i], predicted[i]))

    # Metrics calculation
    val_loss /= len(valid_loader)
    accuracy = 100 * correct / total
    precision = precision_score(true_labels, predictions, average='macro')
    recall = recall_score(true_labels, predictions, average='macro')
    f1 = f1_score(true_labels, predictions, average='macro')

    # Append metrics to lists
    training_losses.append(running_loss / len(train_loader))
    validation_losses.append(val_loss)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    epochs.append(epoch + 1)

    # Print epoch summary
    print(f'Epoch [{epoch + 1}/{num_iterations}], Validation Loss: {val_loss:.3f}, Accuracy: {accuracy:.2f}%')
    print(f'Precision: {precision:.3f}, Recall: {recall:.3f}, F1 Score: {f1:.3f}')

    # Visualization of the confusion matrix
    conf_matrix = confusion_matrix(true_labels, predictions)
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.matshow(conf_matrix, cmap='Blues')
    num_classes = len(classes)
    ax.set_xticks(np.arange(num_classes))
    ax.set_yticks(np.arange(num_classes))
    ax.set_xticklabels(classes)
    ax.set_yticklabels(classes)
    for i in range(num_classes):
        for j in range(num_classes):
            ax.text(j, i, str(conf_matrix[i, j]), ha='center', va='center', color='white')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

    # Visualization of misclassified images at the end of training
    if epoch == num_iterations - 1:
        if len(misclassified_images) > 0:
            fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))
            fig.suptitle('Misclassified Images')
            for idx, (image, true_label, predicted_label) in enumerate(misclassified_images[:9]):
                ax = axes[idx // 3, idx % 3]
                ax.imshow(image.squeeze().cpu().numpy(), cmap='gray')  # Adjust according to your data preprocessing
                ax.set_title(f'True: {classes[true_label]}, Predicted: {classes[predicted_label]}')
                ax.axis('off')
            plt.tight_layout()
            plt.show()

print('Finished Training')
```

```python
# Testing the best model on the test set
cnn_model.eval()
test_loss = 0.0
correct = 0
total = 0
predictions = []
true_labels = []

with torch.no_grad():
    for data in test_loader:
        inputs, labels = data
        outputs = cnn_model(inputs)
        loss_value = loss(outputs, labels)
        test_loss += loss_value.item()

        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        predictions.extend(predicted.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Metrics calculation for test set
test_loss /= len(test_loader)
accuracy = 100 * correct / total
precision = precision_score(true_labels, predictions, average='macro')
recall = recall_score(true_labels, predictions, average='macro')
f1 = f1_score(true_labels, predictions, average='macro')

# Print test set summary
print(f'Test Set, Loss: {test_loss:.3f}, Accuracy: {accuracy:.2f}%')
print(f'Precision: {precision:.3f}, Recall: {recall:.3f}, F1 Score: {f1:.3f}')

# Visualization of the confusion matrix for test set
conf_matrix = confusion_matrix(true_labels, predictions)
fig, ax = plt.subplots(figsize=(8, 8))
ax.matshow(conf_matrix, cmap='Blues')
num_classes = len(classes)
ax.set_xticks(np.arange(num_classes))
ax.set_yticks(np.arange(num_classes))
ax.set_xticklabels(classes)
ax.set_yticklabels(classes)
for i in range(num_classes):
    for j in range(num_classes):
        ax.text(j, i, str(conf_matrix[i, j]), ha='center', va='center', color='white')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Test Set')
plt.show()

# Visualization of misclassified images from test set
misclassified_images_test = []
for i in range(len(true_labels)):
    if true_labels[i] != predictions[i]:
        misclassified_images_test.append((test_loader.dataset[i][0], true_labels[i], predictions[i]))

if len(misclassified_images_test) > 0:
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))
    fig.suptitle('Misclassified Images from Test Set')
    for idx, (image, true_label, predicted_label) in enumerate(misclassified_images_test[:9]):
        ax = axes[idx // 3, idx % 3]
        ax.imshow(image.squeeze().cpu().numpy(), cmap='gray')  # Adjust according to your data preprocessing
        ax.set_title(f'True: {classes[true_label]}, Predicted: {classes[predicted_label]}')
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```
